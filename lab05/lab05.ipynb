Для модели «хищник-жертва»:
$$
\begin{cases}
    \frac{\partial x}{\partial t} = -0.13x(t)+0.041x(t)y(t)
    \\
    \frac{\partial y}{\partial t} = 0.31y(t)-0.042x(t)y(t)
\end{cases}
$$
Постройте график зависимости численности хищников от численности жертв,
а также графики изменения численности хищников и численности жертв при следующих начальных условиях:
$
x_0 = 7, y_0 = 20
$. Найдите стационарное состояние системы.

```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
a = 0.13 # коэффициент естественной смертности хищников
b = 0.31 # коэффициент естественного прироста жертв
c = 0.041 # коэффициент увеличения числа хищников
d = 0.042 # коэффициент смертности жертв
```

```python
def syst2(x, t):
    dx0 = -a*x[0] + c*x[0]*x[1]
    dx1 = b*x[1] - d*x[0]*x[1]
    return dx0, dx1
```

```python
x0 = [7, 20] # начальное значение x и у (популяция хищников и популяция жертв)
```

```python
t = np.arange(0, 400, 0.1)
```

```python
y = odeint(syst2, x0, t)
```

```python
y2 = y[:,1] # массив хищников 
y1 = y[:,0] # массив жертв
```

```python
plt.plot(t,y1, label='хищники')
plt.plot(t,y2, label='жертвы')
plt.legend(loc=1)
plt.grid(axis='both')
```

```python
plt.plot(y1,y2) #построение графика зависимости изменения численности хищников от изменения численности жертв
plt.plot(7,20, 'ro', label='начальное состояние')
plt.plot(b/d,a/c, 'g*', label='стационарное состояние')
plt.title('Зависимость изменения численности')
plt.xlabel('численность жертв')
plt.ylabel('численность хищников')
plt.legend(loc=1)
plt.grid(axis='both')
```
